In Python, numbers are pretty intuitive, but there are two main "flavors" you’ll use 99% of the time, and one special way to handle math that usually surprises beginners.

1. The Two Main Types
Python automatically distinguishes between whole numbers and decimals.

Integers (int): Whole numbers, positive or negative, without a decimal point.

Example: 10, -5, 0

Floats (float): Numbers with a decimal point. Even if it’s .0, Python sees it as a float.

Example: 10.5, -1.2, 7.0

Note: If you add an int and a float (e.g., 5 + 2.0), Python will always promote the result to a float (7.0) to keep the precision.

2. Basic Arithmetic Operations
Python works like a high-powered calculator. Here are the symbols you’ll use:

Addition +
Subtraction -

3. The "Special" Division Operators
This is where new learners often get stuck. Python has two ways to divide:

A. Floor Division (//)
This chops off the decimal and gives you the whole number (it doesn't round; it just discards the remainder).

```python
print(10 // 3)  # Result: 3
```

B. Modulo (%)
This gives you only the remainder left over after division. It is incredibly useful for checking if a number is even or odd.

```python
print(10 % 3)   # Result: 1 (because 3 goes into 10 three times with 1 left over)
```

4. Order of Operations
Python follows the standard math rule of PEMDAS (Parentheses, Exponents, Multiplication/Division, Addition/Subtraction).

```python
result = 10 + 5 * 2    # Result is 20
result = (10 + 5) * 2  # Result is 30
```

5. Helpful Number Shortcuts
Python includes some built-in "cheats" to make life easier:

abs(x): Gives the absolute (positive) value of a number.

round(x, n): Rounds a number to n decimal places.

Underscores for readability: You can use underscores in large numbers to make them readable. Python ignores them!

```python
huge_number = 1_000_000  # Python just sees 1000000
```

In Python, a String is just a sequence of characters. Think of it like beads on a necklace—each bead is a character (a letter, a space, a number, or a symbol), and the string is the whole chain.

Anything wrapped in quotes is a string.

1. Creating Strings
Python is flexible with quotes. You can use single (') or double (") quotes. They do exactly the same thing, but double quotes are handy if your text contains an apostrophe.

```python
name = 'Alice'
message = "I'm learning Python!"  # Double quotes prevent the ' in I'm from breaking the code
```

Triple Quotes (""")
If you need a string to span multiple lines, use three quotes in a row. This is great for long paragraphs or "Docstrings."

```python
long_story = """Once upon a time,
there was a coder who forgot 
to close their quotes..."""
```

2. String Concatenation (The + Sign)
In the "Numbers" lesson, + added values. In "Strings," + glues them together.

```python
first = "Python"
second = "Rocks"
combined = first + " " + second  # Result: "Python Rocks"
```

⚠️ The Common Mistake: You cannot add a string and a number directly. Python will get confused and throw a TypeError.

print("Score: " + 10) ❌ Fail

print("Score: " + str(10)) ✅ Pass (We use str() to turn the number into a string first).

3. F-Strings (The Modern Way)
Instead of messy concatenation with + signs, Python 3.6+ gave us f-strings. They are cleaner, faster, and much more readable. You just put an f before the quotes and use {} for your variables.

```python
name = "Alex"
age = 25
print(f"Hi, my name is {name} and I am {age} years old.")
```

4. Indexing and Slicing
Since a string is a sequence, every character has a specific "address" called an index.
Crucial Rule: Python starts counting at 0, not 1.

```python
word = "Python"

print(word[0])  # Result: "P"
print(word[1])  # Result: "y"
print(word[-1]) # Result: "n" (Negative index starts from the end!)
```

Slicing
You can grab a "slice" of a string using [start:stop]. Note: It includes the start index but stops just before the stop index.

```python
print(word[0:2]) # Result: "Py" (Indices 0 and 1)
```

In programming, an if condition is like a fork in the road. It allows your code to make decisions based on whether something is True or False. If the condition is met, the code runs; if not, the code skips it.

1. The Syntax (The "Wait, Where’s the Bracket?" Rule)
Python uses a colon : and indentation (usually 4 spaces or a Tab) to show which code belongs to the if statement.

```python
is_raining = True

if is_raining:
    print("Take an umbrella!")  # This runs only if is_raining is True
    
print("Have a nice day!")        # This runs no matter what (it's not indented)
```

2. Adding else and elif
Sometimes you want to do one thing if a condition is true, and something else if it isn’t.

if: The first check.

elif (short for "else if"): Use this to check multiple specific conditions in order.

else: The "catch-all" for when none of the above are true.

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C or lower")
```

3. Comparison Operators
To ask Python a question, you need these "Comparison Operators." They always result in a Boolean (True or False).

⚠️ Watch Out: A single = is for assigning a value to a variable. A double == is for comparing two values. This is the most common mistake beginners make!

4. Logical Operators (The "And/Or" Logic)
You can combine multiple conditions into one if statement using and, or, and not.

and: Both conditions must be True.

or: At least one condition must be True.

not: Flips the result (True becomes False).

```python
age = 20
has_ticket = True

if age >= 18 and has_ticket:
    print("Welcome to the concert!")
```

A for loop is used when you want to repeat a block of code a specific number of times or "iterate" over a collection (like a list or a string).

Think of it as an automated way to say: "For every item in this group, do this specific task."

1. The Basic Syntax
The syntax is designed to read almost like a sentence:

```python
# The list we want to look through
friends = ["Joey", "Monica", "Chandler"]

for friend in friends:
    print(f"Hello, {friend}!")
```

What’s happening here?

for: This tells Python we are starting a loop.

friend: This is a temporary variable (you can name it anything). On the first trip through the loop, friend is "Joey". On the second trip, it's "Monica", and so on.

in friends: This tells Python which collection to look at.

: and Indentation: Just like with if statements, the indented code is what actually gets repeated.

2. Using the range() Function
Sometimes you don't have a list; you just want to run code X times. For that, we use range().

```python
for i in range(5):
    print(f"Attempt number: {i}")
```

Note: range(5) generates numbers from 0 to 4. Just like indexing, Python starts at 0 and stops just before the number you provide.

You can also specify a start and end:
range(1, 6) → 1, 2, 3, 4, 5

range(0, 10, 2) → 0, 2, 4, 6, 8 (The third number is the "step"

3. Looping Through Strings
Since strings are just sequences of characters, you can loop through them too!

```python
word = "CODE"
for letter in word:
    print(letter.lower())
```

4. The break and continue Keywords
Sometimes you need to control the loop more strictly:

break: Stop the loop entirely and exit.

continue: Skip the rest of this trip through the loop and jump to the next item.

```python
numbers = [1, 2, 3, 99, 4, 5]

for num in numbers:
    if num == 99:
        break  # The loop stops immediately when it hits 99
    print(num)
```

5. Comparison: When to use if vs for?
if: Checks a condition once.

for: Runs a block of code multiple times (once for every item in a set).

Think of a Function as a "mini-program" inside your code. It’s a block of code that you write once, give a name to, and then "call" whenever you need it.

If you find yourself copying and pasting the same code three times, you should probably put it in a function. In programming, we call this the DRY Principle (Don’t Repeat Yourself).

1. Defining a Function
To create a function, you use the keyword def (short for define).

```python
def say_hello():
    print("Hello there!")
    print("Welcome to Python.")

# Calling the function (actually running it)
say_hello()
say_hello()
```

Note: Just like if and for, the code inside the function MUST be indented.

2. Parameters (Input)
Functions are much more useful when you can send them data to work with. These "placeholders" are called parameters.

```python
def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Alex")  # Result: Hello, Alex!
greet_user("Sam")   # Result: Hello, Sam!
```

3. The return Keyword (Output)
This is the part that usually confuses beginners.

print() just shows something on the screen.

return gives a value back to the rest of the program so you can save it or use it later.

Think of it like a pizza delivery: print() is the delivery guy showing you the pizza through the window. return is the delivery guy actually handing you the pizza so you can eat it.

```python
def add_numbers(a, b):
    return a + b

# We can save the result of the function into a variable!
total = add_numbers(10, 5)
print(total) # Result: 15
```

4. Default Parameters
You can give a parameter a "fallback" value in case the user doesn't provide one.

```python
def power_up(name, level=1):
    print(f"{name} is now level {level}!")

power_up("Mario", 10) # Mario is now level 10!
power_up("Luigi")      # Luigi is now level 1! (Uses the default)
```

We’ve covered Lists (which are like flexible shelves), but sometimes you want data that is "locked in" or data that works like a real-world filing system. That’s where Tuples and Dictionaries come in.

1. Tuples: The "Locked" List
A Tuple is almost exactly like a list, but with one major catch: it cannot be changed. Once you create it, you can't add, remove, or swap items. In tech terms, they are immutable.

Syntax: Use parentheses () instead of square brackets.

```python
# A tuple of coordinates
location = (40.7128, 74.0060)

# You can access items just like a list
print(location[0])  # Result: 40.7128

# Attempting to change it will cause an error!
# location[0] = 50.0  <-- This would crash your program
```

Why use a Tuple?
Safety: Use them for data that shouldn't change (like GPS coordinates, RGB color codes, or a person's birthday).

Speed: Python processes tuples slightly faster than lists.

2. Dictionaries: The "Key-Value" Store
A Dictionary (or dict) is arguably the most powerful data structure in Python. Instead of using numbers (0, 1, 2) to find items, you use Keys.

Think of it like a real dictionary: you look up a Word (the Key) to find its Definition (the Value).

Syntax: Use curly braces {} and a colon : to separate keys from values.

```python
# A dictionary representing a car
car = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

# Accessing data using the Key
print(car["brand"])  # Result: "Ford"
```

Managing Dictionaries
Dictionaries are mutable, so you can update them on the fly:

```python
# Adding a new key-value pair
car["color"] = "Red"

# Updating an existing value
car["year"] = 2024

# Removing a pair
del car["model"]
```

4. Useful Dictionary Methods
When working with dictionaries, you'll often want to extract just the labels or just the data:

.keys(): Gives you a list of all keys.

.values(): Gives you a list of all values.

.items(): Gives you both (useful for looping!).

```python
user = {"username": "coder123", "status": "online"}

for key, value in user.items():
    print(f"The {key} is {value}")
```

Think of Modules as pre-packed toolboxes and Pip as the delivery service that brings new toolboxes to your door.

As you get better at Python, you won’t want to write everything from scratch. Instead, you’ll use code that other people (or the Python creators) have already written.

1. What is a Module?
A module is simply a file containing Python code (functions, variables, or classes).

There are three types:

Built-in Modules: These come pre-installed with Python (the "Standard Library").

Your Own Modules: Other .py files you create in the same folder.

External Modules: Libraries written by the community (like for AI, web scraping, or games).

2. How to Use Modules (import)
To use a module, you have to "invite" it into your script using the import keyword.

A. The Standard Import

```python
import math

print(math.sqrt(16))  # Result: 4.0
```

B. Importing Specific Parts
If you only need one specific tool, you can grab just that one to save memory and typing.



```python
from random import randint

print(randint(1, 10))  # Generates a random number between 1 and 10
```

C. Giving a Nickname (as)
Some modules have long names. We often give them short aliases to save our fingers from typing.

```python
import datetime as dt

print(dt.datetime.now())
```

3. What is Pip?
Pip stands for "Pip Installs Packages." It is the Package Manager for Python.

If Python were a smartphone, Pip would be the App Store. It allows you to download and install over 500,000+ external libraries from the Python Package Index (PyPI).

Note: Pip is a Command Line tool. You do not type pip commands inside your Python script; you type them in your Terminal, Command Prompt, or VS Code Terminal.

4. Common Pip Commands
You’ll use these in your terminal to manage your external "toolboxes."

Example:
If you wanted to write code that sends HTTP requests to websites, you would run this in your terminal:
pip install requests

Then, in your Python file, you could use it:

```python
import requests

response = requests.get("https://www.google.com")
print(response.status_code) # Result: 200 (which means it worked!)
```

When you share your code with a friend, you don't send them the actual external libraries (which are huge). Instead, you send a text file called requirements.txt that lists everything they need to install.

File handling is where your programs start to feel "real." Until now, every time you stopped your Python script, your data disappeared. File handling allows you to save data to a text file (long-term storage) so you can load it back later.

Think of Python as an office worker and a .txt file as a folder in a filing cabinet. To do anything, the worker has to open the cabinet, do the work, and then—most importantly—close it.

1. The open() Function and Modes
To work with a file, you use the open() function. It takes two main arguments: the filename and the mode.

2. The "Right Way": The with Statement
In the old days, you had to manually close files using file.close(). If your program crashed before that line, the file could get corrupted.

Modern Python uses the with keyword. It’s a "context manager" that automatically closes the file for you, even if things go wrong.

3. Reading from a File
Let's say you have a file named notes.txt. Here is how you get the data out:

```python
with open("notes.txt", "r") as file:
    for line in file:
        print(line.strip()) # .strip() removes the extra newline (\n)
```

4. Writing and Appending
When you write to a file, Python will create the file if it doesn't exist yet.

Writing (Overwriting):

```python
with open("goals.txt", "w") as file:
    file.write("Learn Python\n")
    file.write("Build an app\n")
# If goals.txt already had text, it's gone now!
```

Appending (Adding):

```python
with open("goals.txt", "a") as file:
    file.write("Get a coding job\n")
# This adds the line to the bottom of the existing list.
```

5. Common "Gotchas"
The Newline Character (\n): Python doesn't automatically add a new line when you use .write(). You have to manually add \n at the end of your string, or everything will be bunched up on one line.

File Paths: If your file is in a different folder, you need to provide the path (e.g., "data/notes.txt").

FileNotFoundError: If you try to 'r' (read) a file that doesn't exist, Python will throw an error.
